# <u>PRIME NUMBERS</u>

```python
num=int(input("Enter a number:"))
if(num<=1):
    print(f"{num} is not a prime number.")
elif(num==2):
    print(f"2 is a prime number.")
elif(num%2==0):
    print(f"{num} is not a prime number.")
else:
    print(f"{num} is a prime number.")
```

# <u>PRODUCT OF RANDOM NUMBERS</u>

```python
import random as rd
a=rd.randint(1,50) #generates a random integer number between the range 1 to 50
b=rd.randint(1,50)
print(f"What should be the product of {a} and {b}?")
ans=int(input("Product : ")) #asking user to input the product of two nos.
print(f"Answer: {ans}")
if(ans==(a*b)):
    print(f"Well Done!You are absolutely right.The answer was {a*b}.")
else:
    print(f"Oops!You are wrong.The answer was {a*b}.")

```

# <u>SQUARES OF EVEN & ODD NUMBERS(100-200)</u>

```python
import numpy as np
import pandas as pd
even=[]#empty list for even nos
odd=[]#empty list for odd nos
sq_even=[]#empty list for square of even nos
sq_odd=[]#empty list for square of odd nos
for i in range(100,202):
    if(i%2==0):
        even.append(i)#add even nos to the even list
        sq_even.append(i**2)#adds square of even nos to the sq_even list
    else:
        odd.append(i)#add odd nos to the even list
        sq_odd.append(i**2)#adds square of odds nos to the sq_odd list

pd.DataFrame({
    "Even_no":even,
    "SquaresEven":sq_even,
    "Odd_no":odd,
    "OddSquares":sq_odd
})#prints all the nos and their square in a tabular format

#here we have considered an odd no(201) just to equal the length of both even and odd number list
```

# <u>PALINDROME</u>

```python
#function to check palindrome
def is_palindrome(str):
    char_list=list(str)#creates a list of characters in the string
    ch=char_list.reverse()#reverses the list
    rev_str="".join(char_list)#converts the reversed list into a string
    print(f"Reversed String: {rev_str}")
    #check if the input string  is same as the reversed string
    if(str==rev_str):
        print(f"'{str}' is a palindrome")
    else:
        print(f"'{str}' is not a palindrome")

string=input("Enter a string: ")
print(f"Input: {string}")
is_palindrome(string)
```
